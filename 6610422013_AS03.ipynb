##Import library

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Pomaccel/DADS5001-Data-Analytics-and-Data-Science-Tools-and-Programming/main/student_spending%20.csv")
df.head()
```

##Check size

```python
df.shape
```

```python
df.info()
```

```python
df = df.rename(columns={"preferred_payment_method": "payment_method", "food": "food ($)"})
df.head()
```

## Check unique by each columns

```python
check_unique = df.columns[1:5]

for i in check_unique :
  print(i,end = '')
  print( df[i].unique())
```

```python
fig = px.bar(df,
             x='year_in_school',
             y='food ($)',
             #size='food',
             color = 'payment_method',
             facet_col = 'major',
             facet_row = 'payment_method',
             title = 'Student spending for food (in dollars) by major from 1000 students by each Year in School',
             width = 2000,
             height = 1000,)
fig.show()
```
